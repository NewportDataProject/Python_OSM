```python
import osmnx as ox
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib.cm as cm
print(ox.__version__)
print(nx.__version__)
```

```python
# define the colormap
cmap = cm.get_cmap('PuRd')
```

```python
# get and show city bounds, for fun
city = ox.gdf_from_place('Newport, RI')
ox.plot_shape(ox.project_gdf(city))
```

```python
# get driving network
streets = ox.graph_from_place('Newport, RI', network_type='drive')
```

```python
btwn_streets = nx.betweenness_centrality(streets)
```

```python
print(max(btwn_streets.values()))
print(sum(btwn_streets.values())/len(btwn_streets))
```

```python
# plot graph illustrating betweenness
ox.plot_graph(streets, fig_height=16, edge_color=cmap(list(btwn_streets.values())), edge_linewidth=2.5, node_size=0)
```

```python
ec = ['r' if data['oneway'] else 'b' for u, v, key, data in streets.edges(keys=True, data=True)]
ox.plot_graph(streets, node_size=0, edge_color=ec,fig_height=16)
```
